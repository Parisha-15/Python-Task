```python
# Multilevel
class A:
    def __init__(self):
#         super().__init__() No error will be triggered
        print("Constructor A")
    def show(self):
        print("A")
class B(A):
    def __init__(self):
        print("Constructor B")
        super().__init__()
    def show(self):
        print("B")
class C(B):
    def __init__(self):
        super().__init__()
        print("Constructor C")
    def show(self):
        print("C")
c=C()
c.show()
print(C.__mro__)
C.__mro__ # here C is class name 
```

```python
# Multiple Inheritance
class A:
    def __init__(self):
        print("Constructor A")
    def show(self):
        print("A")
class B(A):
    def __init__(self):
        print("Constructor B")
        super().__init__()
    def show(self):
        print("B")
class C(B,A):
    def __init__(self):
        super().__init__()
        print("Constructor C")
    def show(self):
        super().show()
        print("C")
        
        
        
        
        
c=C()
c.show()
print(C.__mro__)
C.__mro__
```

```python
class Spy:
    def __init__(self,mark):
        self.mark=mark
    def __add__(self,other):
        print(self.mark-other.mark)
    def __str__(self):
        return("Hello")
    def __truediv__(self,other):
        print(self.mark-other.mark)
    def __ge__(self,other):
        print(self.mark-other.mark)
    def __le__(self,other):
        print(self.mark-other.mark)
    def __sub__(self,other):
        print(self.mark-other.mark)
   
    def __mul__(self,other):
        print(self.mark-other.mark)
    def __mod__(self,other):
        print(self.mark-other.mark)
    def __pow__(self,other):
        print(self.mark-other.mark)
    def __eq__(self,other):
        print(self.mark-other.mark)
    def __ne__(self,other):
        print(self.mark-other.mark)
    def __lt__(self,other):
        print(self.mark-other.mark)
    def __gt__(self,other):
        print(self.mark-other.mark)
   
   
s1=Spy(20)
s2=Spy(100)
s1+s2
print(s1)
s1-s2
s1/s2
```
