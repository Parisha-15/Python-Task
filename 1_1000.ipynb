```python
n=int(input("Enter num: "))
def Demo(n):    
    d={1:"ONE",2:"TWO",3:"THREE",4:"FOUR",5:"FIVE",6:"SIX",7:"SEVEN",8:"EIGHT",9:"NINE",10:"TEN",11:"ELEVEN",12:"TEWELVE",
      13:"THIRTEEN",14:"FOURTEEN",15:"FIFTEEN",16:"SIXTEEN",17:"SEVENTEEN",18:"EIGHTEEN",19:"NINETEEN",20:"TWENTY",30:"THIRTY",
      40:"FORTY",50:"FIFTY",60:"SIXTY",70:"SEVENTY",80:"EIGHTY",90:"NINETY"}
    for i in range(100):
        if n<=20:
            return d[n]
        if n<100:
            if n%10==0:
                return d[n]
            else:
                return d[(n//10)*10]+d[n%10]
        if n<1000:
            if n%100==0:
                return d[n//100]+"HUNDERED"
            else:
                return d[n//100]+"HUNDRED"+Demo(n%100)
z=Demo(n)
print(z)
    
    
```
