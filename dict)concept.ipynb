```python
name=["A","B","C","D"]
marks=[10,20,30,40]
z1=zip(name,marks)
print(z1)
z=dict(zip(name,marks))
print(z)
print(z.get("A"))
print(z.get("AAA"))
print(z.get("AAA",-1))
for i in z.keys():
    print(i,end=",")
for i in z.keys():
    print(z[i],end=",")

```
