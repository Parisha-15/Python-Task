```python
# Armstrong Number  eg: 153 i.e. 1^3 + 5^3 + 3^3 ==> 1 + 125 + 27 ==> 153
c=0
ans=0
n=int(input("Enter your number"))
z=n   # to print the input number
t=n   # for counting of digit for power purpose 
while t>0:
    t=t//10
    c=c+1
while n>0:
    d=n%10
    ans=ans+d**c
    n=n//10
if z==ans:
    print(z,"is an Armstrong number")
else:
    print(z,"is not an Armstrong number")
```
