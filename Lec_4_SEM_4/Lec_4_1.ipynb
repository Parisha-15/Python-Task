```python
import pandas as pd
import numpy as np
df=pd.read_csv("auto-mpg.csv")
df
```

```python
df=df[df["horsepower"]!="?"]
```

```python
df
```

```python
df["horsepower"]=df["horsepower"].astype("int")
```

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
import seaborn as sns
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

# vmin used for value highlight and minimum value mate colour code used for weather forecast

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=0)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=0.5)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=0.5,linewidths=5,linecolor="cyan")
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=0.5,linewidths=5,linecolor="cyan",cbar=False)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=0.5,linewidths=5,linecolor="cyan",cbar=True)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=0.5,linewidths=5,linecolor="cyan",cbar=True,cmap="coolwarm")
```
