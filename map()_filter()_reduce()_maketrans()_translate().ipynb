```python
L=[1,2,3,4,5,6,7,8]
z=list(filter(lambda x:x%2==0,L))
print(z)
z=set(map(lambda x:x+2,z))
print(z)

from functools import*
z=reduce(lambda x,y:x*y,z)
print(z)
```
