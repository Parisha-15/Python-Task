```python
x="Happy New Year"
y=len(x)   
print(y)
print(x[1:5:1])
print(x[1:5:3])
print(x[1:5:4])
print(x[1:5:-1]) # Blank string is o/p
print(x[1:-1])
print(x[-3:-8:1]) # Blank string is o/p
print(x[-3:-9:-1])
print(x[::])  # shows that by default increment of 1 
# here 1 value ==> start , 2 value ==> end , 3 value ==> increment/decrement
print(x[::1])
print(x[::-1])  # ====  Reverse String logic
print(x[0:len(x)])
print(x[0:len(x):len(x)])





```
