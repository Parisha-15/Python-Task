# inner join gives common 
# while outer join gives all data

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"C":[5,7],"B":[7,9]})
```

```python
pd.concat((df1,df2),join="inner")
```

```python
pd.concat((df1,df2),join="outer")
```

```python
df1=pd.DataFrame({"B":[1,2],"A":[3,4]})
df2=pd.DataFrame({"C":[5,7],"B":[7,9]})
```

```python
pd.concat((df1,df2),join="inner")
```

```python
pd.concat((df1,df2),join="outer")
```

```python
pd.concat((df1,df2),join="outer",sort=True)
```

# concat & join 3M or MCQ 

```python
df1=pd.DataFrame({"sub":["FSD","Python","DM"],"fac":["nas","vha","mss"],"id":[1,2,4]})
df2=pd.DataFrame({"id":[1,2,3],"mark":[9,10,8]})
```

```python
df1
```

```python
df2
```

# for merge one column same is the requirement

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
df1.merge(df2,on="id")
```

# In terms of sql df1 ==> left table and df2 ==> right table

```python
df1.merge(df2,on="id",how="left")
```

```python
df1.merge(df2,on="id",how="right")
```

```python
df1.merge(df2,how="cross")  
#  how ==> cross gives possible combination , will work in other version
# student table and subject table ma common subject ave ena mate use karvanu cross
```

```python
df1=pd.DataFrame({"sub":["FSD","Python","DM"],"fac":["nas","vha","mss"],"id_x":[1,2,4]})
df2=pd.DataFrame({"id_v":[1,2,3],"mark":[9,10,8]})
```

```python
df1.merge(df2,left_on="id_x",right_on="id_v")
```

```python
df1.merge(df2,left_on="id_x",right_on="id_v",left_index=True,right_index=True)
# banne ma same index hovi joi ae
```
