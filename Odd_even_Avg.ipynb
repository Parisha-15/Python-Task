```python
s1=0
c1=0
s2=0
c2=0
while True:
        n=input("Enter number or press Q/q for exit")
        if n=='q' or n=='Q':
            break
        else:
            n=int(n)
            if n%2==0:
                s1=s1+n
                c1=c1+1
            elif n%2!=0:
                s2=s2+n
                c2=c2+1
print(f"sum {s1} and count {c1} and avg {s1/c1}")
print(f"sum {s2} and count {c2} and avg {s2/c2}")

            
            
```
