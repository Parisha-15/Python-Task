```python
#index hai
x="Dhruv Dhoni"
z=x.find("D")
if z==-1:
    print(f"{z} not found")
else:
    print(f"{z}")
```

```python
#pos hai
x="Dhruv Dhoni"
z=x.find("D")
if z==-1:
    print(f"{z} not found")
else:
    print(f"{z+1}")
```
